```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
df = pd.read_csv('zomato.csv',encoding='latin-1')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final = pd.merge(df,df_country,on='Country Code', how = 'left')
final
```

```python
final.columns
```

```python
final.Country.value_counts()  #india transaction is highest
```

```python
country_names = final.Country.value_counts().index  # show only country name according to highest value
country_names
```

```python
country_values = final.Country.value_counts().values  # show only values of country
country_values
```

```python
## Pie Chart- Top 5 countries that uses zomato

plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.1f%%')
plt.title('Top 3 countries that uses zomato')
plt.show()
```

**Observation:Zomato maximum records or transaction are from India After that USA and then United Kingdoms**

```python
final.columns
```

```python
rating = final.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0 : 'Rating count'})
rating
```

**Observation**
1. When Rating is between 4.5 to 4.9---> Excellent
2. When Rating are between 4.0 to 3.4--->very good
3. when Rating is between 3.5 to 3.9----> good
4. when Rating is between 3.0 to 3.4----> average
5. when Rating is between 2.5 to 2.9----> average
6. when Rating is between 2.0 to 2.4----> Poor

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x='Aggregate rating',y='Rating count',data=rating)
plt.title('Rating wise rating count')
plt.xlabel('Aggregate rating')
plt.ylabel('Rating count')
plt.show()
```

```python
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x='Aggregate rating',y='Rating count',hue= 'Rating color', data=rating)
plt.title('Rating wise rating count')
plt.xlabel('Aggregate rating')
plt.ylabel('Rating count')
plt.show()
```

Observation:

1. Not Rated count is very high
2. Maximum number of rating are between 2.5 to 3.4

```python

sns.countplot(x='Rating color',data=rating, palette=['blue','red', 'orange','yellow','green','violet'])
```
